```python
import pandas as pd
import numpy as np

df = pd.read_csv('archive/Combined_dataset.csv')
print("Loaded successfully!")
print(df.head())
```

```python
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nData Types:\n", df.dtypes)
print("\nFirst 5 rows:\n", df.head())
print("\nLast 5 rows:\n", df.tail())
print("\nBasic Stats:\n", df.describe())
```

```python
print("Missing values:\n", df.isnull().sum())

# Fill numeric columns with mean
df.fillna(df.select_dtypes(include='number').mean(), inplace=True)

# Drop remaining nulls
df.dropna(inplace=True)

print("\nAfter cleaning:\n", df.isnull().sum())
```

```python
# Show available columns first
print("Columns:", df.columns.tolist())

# Filter rows (example: where ratings > 3)
filtered = df[df.select_dtypes(include='number').columns[0]] > 0
print("Filtered rows:\n", df[filtered].head())

# Select first 3 columns
selected = df.iloc[:, :3]
print("\nSelected columns:\n", selected.head())
```

```python
print("Duplicates before:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("Duplicates after:", df.duplicated().sum())
```

```python
# Check column names
print(df.columns.tolist())

# Create total_amount = price * quantity (adjust column names if different)
# Common column names in this dataset:
if 'price' in df.columns and 'quantity' in df.columns:
    df['total_amount'] = df['price'] * df['quantity']
elif 'Price' in df.columns and 'Quantity' in df.columns:
    df['total_amount'] = df['Price'] * df['Quantity']
else:
    # Use first two numeric columns
    num_cols = df.select_dtypes(include='number').columns
    df['total_amount'] = df[num_cols[0]] * df[num_cols[1]]

print("total_amount column added:\n", df['total_amount'].head())
```

```python
df.to_csv('cleaned_shopping_data.csv', index=False)
print("✅ Cleaned file saved as cleaned_shopping_data.csv!")
```
